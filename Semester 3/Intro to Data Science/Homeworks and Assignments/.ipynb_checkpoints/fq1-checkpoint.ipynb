```python
data = []
result = []
with open("Assignment1_data\Assignment-1 data\ex1data1.txt",'r') as data_file:
    for i in data_file:
        temp=i.split('\n')[0].split(',')
        data.append(float(temp[0]))
        result.append(float(temp[1]))

print(data)
print("-----------")
print(result)
```

```python
# plotting the data

import matplotlib.pyplot as plt

plt.scatter(data, result,marker='x',color='red')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.suptitle('Figure 1: Scatter plot of training data');
```

```python
theeta1 = 0
theeta2 = 1
alpha = 0.5
hx = 0
der_cost = 0
summation = 0
cost = 0
cost_prev = 0
size = len(data) # m in formula


while(cost != 0):
    for i in range(0,size): # cost function
        hx = theeta1 + (theeta2 + data[i])
        summation += (hx - result[i])**2
        der_cost += hx - result[i]
    cost = summation / (2*size)
    der_cost /= size
    cost_prev = cost
    if(cost > 0): # gradient descent
        theeta1 = theeta1 - alpha * der_cost * data[i]
        theeta2 = theeta2 - alpha * der_cost * data[i]

print(cost)
print(theeta1)
print(theeta2)
```

```python
# plotting the data

import matplotlib.pyplot as plt
y=[] # an array to store points of linear regression
for i in data: # calculating the points
    y.append(theeta1+(theeta2*i))

plt.scatter(data, result , marker='x', color='red')
plt.plot(data ,y) # plotting a custom line by two values of x
plt.legend(['Linear regression','Training data'], loc="lower right")
plt.xlabel('Population of City in 10,000s')
plt.ylabel('profit in $10,000s')
plt.suptitle('Figure 2: Training data with linear regression fit');
```
