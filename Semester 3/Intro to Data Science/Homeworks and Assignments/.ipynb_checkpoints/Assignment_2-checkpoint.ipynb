```python
def sigmoid_function(theetas, X):
    theetas = np.array(theetas)
    transpose_theeta = np.transpose(theetas)
    dot_product = np.dot(transpose_theeta, X)
    g = 1 / (1 + (math.e)**-(dot_product))
    return g
```

```python
m, n = x.shape
x = np.concatenate([np.ones((m, 1)), x], axis=1)
theeta_intial = np.ones(n+1)
```

```python
def cost_function(theta, X, y):
    J = 0
    grad = np.zeros(len(theta))
    for i in range(len(y)):
        h = sigmoid_function(theta, X[i])   
        J += -y[i]*math.log(h)-(1-y[i])*math.log(1-h)

    J *= 1/m

    return J, grad

theeta_initial = [-25.16131856, 0.20623159, 0.20147149]
j= cost_function(theeta_initial, x, y)

print("Intial Theetas:", theeta_initial)
print("Cost J:", j)
print("Optimal theetas:", theeta_initial)
```

```python
def plotData(x, y):
    positive = y == 1 #Postive will be the array where indices of positive examples are stored.
    negative = y == 0    #Negative will be the array where indices of Negative examples are stored.

    plt.figure(figsize=(8, 6))

    plt.plot(x[positive, 0], x[positive, 1], 'k+')
    plt.plot(x[negative, 0], x[negative, 1], 'ko', mfc='y')

    plt.xlabel('Exam 1 score')
    plt.ylabel('Exam 2 score')
    plt.title('Figure 1: Scatter plot of training data')
    plt.legend(['Admitted', 'Not admitted']) 
    
plotData(x, y)
```

```python
def mapFeature(x1,x2,degree):
    """
    take in numpy array of x1 and x2, return all polynomial terms up to the given degree
    """
    out = np.ones(len(x1)).reshape(len(x1),1)
    for i in range(1,degree+1):
        for j in range(i+1):
            terms= (x1**(i-j) * x2**j).reshape(len(x1),1)
            out= np.hstack((out,terms))
    return out
X = mapFeature(x[:,0], x[:,1],6)
```

```python
def sigmoid(z):
    """
    return the sigmoid of z
    """
    return 1/ (1 + np.exp(-z))
    
def costFunctionReg(theta, X, y ,Lambda):
    """
    Take in numpy array of theta, X, and y to return the regularize cost function and gradient
    of a logistic regression
    """
    
    m=len(y)
    y=y[:,np.newaxis]
    predictions = sigmoid(X @ theta)
    error = (-y * np.log(predictions)) - ((1-y)*np.log(1-predictions))
    cost = 1/m * sum(error)
    regCost= cost + Lambda/(2*m) * sum(theta**2)
    
    # compute gradient
    j_0= 1/m * (X.transpose() @ (predictions - y))[0]
    j_1 = 1/m * (X.transpose() @ (predictions - y))[1:] + (Lambda/m)* theta[1:]
    grad= np.vstack((j_0[:,np.newaxis],j_1))
    return regCost[0], grad
```

```python
# Initialize fitting parameters
initial_theta = np.zeros((X.shape[1], 1))
# Set regularization parameter lambda to 1
Lambda = 1
#Compute and display initial cost and gradient for regularized logistic regression
cost, grad=costFunctionReg(initial_theta, X, y, Lambda)
print("Cost at initial theta (zeros):",cost)
```

```python
def gradientDescent(X,y,theta,alpha,num_iters,Lambda):
    """
    Take in numpy array X, y and theta and update theta by taking num_iters gradient steps
    with learning rate of alpha
    
    return theta and the list of the cost of theta during each iteration
    """
    
    m=len(y)
    J_history =[]
    
    for i in range(num_iters):
        cost, grad = costFunctionReg(theta,X,y,Lambda)
        theta = theta - (alpha * grad)
        J_history.append(cost)
    
    return theta , J_history
theta , J_history = gradientDescent(X,y,initial_theta,1,800,1)
# print("The regularized theta using ridge regression:\n",theta)
```

```python
def mapFeaturePlot(x1,x2,degree):
    """
    take in numpy array of x1 and x2, return all polynomial terms up to the given degree
    """
    out = np.ones(1)
    for i in range(1,degree+1):
        for j in range(i+1):
            terms= (x1**(i-j) * x2**j)
            out= np.hstack((out,terms))
    return out

positive = y == 1
negative = y == 0

plt.figure(figsize=(8,6))

plt.plot(x[positive, 0], x[positive, 1], 'k+')
plt.plot(x[negative, 0], x[negative, 1], 'ko', mfc='y')

# Plotting decision boundary
u_vals = np.linspace(-1,1.5,50)
v_vals= np.linspace(-1,1.5,50)
z=np.zeros((len(u_vals),len(v_vals)))
for i in range(len(u_vals)):
    for j in range(len(v_vals)):
        z[i,j] =mapFeaturePlot(u_vals[i],v_vals[j],6) @ theta
plt.contour(u_vals,v_vals,z.T,0)
plt.xlabel("Exam 1 score")
plt.ylabel("Exam 2 score")
plt.title('Figure 4: Training data with decision boundary ( λ = 1 )')
plt.legend(['y = 1', 'y = 0']);
```
